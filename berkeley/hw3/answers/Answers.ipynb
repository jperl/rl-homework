```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
```

# Part 1

## Question 1

```python
def plot_values(data, mean_label=None, max_label=None):
    df = pd.DataFrame(data)
    mean_r = df.rolling(100).mean()
    max_r = mean_r.cummax()
    
    if mean_label:
        plt.plot(mean_r, label=mean_label)
    
    if max_label:
        plt.plot(max_r, label=max_label)

def show_plot():
    plt.legend()
    plt.xlabel("Episode")
    plt.ylabel("Score")
    plt.show()
```

```python
default = pickle.load(open("./default.pkl", "rb"))[:2776]
plot_values(default, "Mean 100-ep Reward", "Best Mean Reward")
show_plot()
```

## Question 2

```python
double_q = pickle.load(open("./double_q.pkl", "rb"))[:2776]
plot_values(default, max_label="Best Mean Reward")
plot_values(double_q, max_label="Best Double-Q Reward")
show_plot()
```

## Question 3

```python
one_512 = pickle.load(open("./default.pkl", "rb"))[:2150]
plot_values(one_512, max_label="1x512")

two_512 = pickle.load(open("./2x512.pkl", "rb"))[:2150]
plot_values(two_512, max_label="2x512")

three_512 = pickle.load(open("./3x512.pkl", "rb"))[:2150]
plot_values(three_512, max_label="3x512")

one_1024 = pickle.load(open("./1x1024.pkl", "rb"))[:2150]
plot_values(one_1024, max_label="1x1024")

two_1024 = pickle.load(open("./2x1024.pkl", "rb"))[:2150]
plot_values(two_1024, max_label="2x1024")

three_1024 = pickle.load(open("./3x1024.pkl", "rb"))[:2150]
plot_values(three_1024, max_label="3x1024")

show_plot()
```

The second 512 layer noticeably improves the q-network performance. Increasing the size to 1024 units improves the convergence speed, but does not improve the score in the long term over the 2x512 network.
