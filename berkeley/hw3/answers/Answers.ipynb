```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
```

# Question 1

```python
data = pickle.load(open("./pong.pkl", "rb"))[0:2000]
df = pd.DataFrame(data)
mean_r = df.rolling(100).mean()
max_r = mean_r.cummax()

rmean, = plt.plot(mean_r, label="Mean 100-ep Reward")
rmax, = plt.plot(max_r, label="Best Mean Reward")
plt.legend("top left", handles=[rmean, rmax])
plt.xlabel("Episode")
plt.ylabel("Score")
plt.show()
```
